# Title Basics

```python
df = pd.read_csv('Data/title_basics.csv.gz')
df.info()
df.head()
```

## Genre

### OHE genres column

```python
gen_loc = df.loc[0,'genres']
print(type(gen_loc))
gen_loc
```

```python
#convert string to list
df['genres_split'] = df['genres'].str.split(",")
df[:3]
```

```python
exploded_genres = df.explode('genres_split')
exploded_genres
```

```python
unique_genres = sorted(exploded_genres['genres_split'].unique())
unique_genres
```

```python
title_genres = exploded_genres[['tconst','genres_split']].copy()
title_genres.head()
```

```python
## Making the genre mapper dictionary
genre_ints = range(len(unique_genres))
genre_map = dict(zip(unique_genres, genre_ints))
genre_map
```

```python
## make new integer genre_id and drop string genres
title_genres['genre_id'] = title_genres['genres_split'].map(genre_map)
title_genres = title_genres.drop(columns='genres_split')
```

```python
title_genres.head()
```

```python
genres = pd.DataFrame({'Genre_Name':genre_map.keys(),
                            'Genre_ID':genre_map.values()})
genres.head()
```

## Title Basics

```python
title_basics.head()
```

```python
## Calculate max string lengths for object columns
key_len = title_basics['Movie ID (tconst)'].fillna('').map(len).max()
title_len = title_basics['Primary Title'].fillna('').map(len).max()
## Create a schema dictonary using Sqlalchemy datatype objects
df_schema = {
    "Movie ID": String(key_len+1), 
    "Primary Title": Text(title_len+1),
    'Start Year':Float(),
    'End Year':Float(),
    'Runtime':Integer()}
```

```python
title_basics.to_sql('title_basics',engine,dtype=df_schema,if_exists='replace',index='False')
```

```python
engine.execute('ALTER TABLE title_basics ADD PRIMARY KEY (`Movie ID`);')
```
